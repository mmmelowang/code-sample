Data preprocessing

a)Convert categorical features into ordinal values

```python
from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder
data=df.copy()
ordinal_encoder = OrdinalEncoder()
data.iloc[:,1:] = ordinal_encoder.fit_transform(data.iloc[:,1:]) #convert features except gender,which should be converted into dummy variable.
data.head()
```

```python
data.describe()
```

b)split data in training and testing sets


```python
from sklearn.model_selection import train_test_split

y = data["Separate"] # Assign outcome variable,the one I want to predict
X = data.loc[:,data.columns != "Separate"] #Assign all other values except ID as X-values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)  

X_train.head()
```

Selecting model

a) Decision trees-Gradient boosted classification 

```python
from sklearn.datasets import make_hastie_10_2 
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(n_estimators=1000, learning_rate=.01,
    max_depth=1, random_state=0).fit(X_train, y_train)
clf.score(X_test, y_test) 
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, y_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
#we can extract important features:
clf.feature_importances_ #It seems that gender,household income and occupation are more important features.
```

b)SVM

Before doing SVM, let's first use PCA to filter noise and feature extraction to reduce data dimension.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(X)
X_pca=pca.transform(X)

print(pca.explained_variance_) # explained variance for for each principal component vect
```

```python
#the two principal components
df_plot = pd.DataFrame(pca.fit_transform(X), columns=['PC1', 'PC2'], index=X.index)
df_plot
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1],
            c=y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()# see the variance in a 2D plot
```

```python
#split train and test data using X_pca
from sklearn.model_selection import train_test_split

X_pca_train, X_pca_test, y_train, y_test = train_test_split(X_pca, y, random_state=42) 
```

```python
from sklearn import svm
#create classifier
svm=svm.SVC(kernel='linear')
svm.fit(X_pca_train,y_train)
y_pred=svm.predict(X_pca_test)

from sklearn import metrics
print("accuracy:",metrics.accuracy_score(y_test,y_pred))
```

3)Logistic regression model

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")

best_ting = {
    'max_iter':[20,40,60],
    'C':[0.01,0.1,1,10],'penalty':['l2']}
best_g = GridSearchCV(LogisticRegression(),best_ting,cv=5)
best_g.fit(X_train, y_train)

print(best_g.best_params_)
```

```python
logreg=LogisticRegression(C=0.01,max_iter=20,penalty='l2').fit(X_train, y_train)
logreg.score(X_test, y_test)
print(logreg.score(X_test, y_test)))
```

4)KNeighborsClassifier

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

param_grid = {'n_neighbors': np.arange(1, 15, 2)}

grid = GridSearchCV(KNeighborsClassifier(), param_grid=param_grid, cv=10)

grid.fit(X_train, y_train)
print("best mean cross-validation score: {:.3f}".format(grid.best_score_))
print("best parameters: {}".format(grid.best_params_))
print("test-set score: {:.3f}".format(grid.score(X_test, y_test)))
```
